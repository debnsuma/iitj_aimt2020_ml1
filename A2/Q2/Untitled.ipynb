```python
#Importing modules
from sklearn.ensemble import AdaBoostClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.ensemble import VotingClassifier

import sys
```

```python
dataset = pd.read_csv("heart.csv")
X=dataset.drop(['target'], axis=1)
Y=dataset['target']
```

```python
X_train, X_test, Y_train, Y_test= train_test_split( X, Y, test_size=0.2, random_state=42)
model=DecisionTreeClassifier( criterion='entropy', max_depth=1, random_state=42)
```

```python
#Range of n_estimator values for part a of the question
parameter_range=np.arange(20,260,20)
```

```python
#caulculating train errors for all values of n and storing in train_result
train_result=[]
z=list(range(12))
for count1 in z:
    i=parameter_range[count1]
    AdaBoost_1=AdaBoostClassifier(n_estimators=parameter_range[count1], learning_rate=1)
    boostmodel=AdaBoost_1.fit( X_train, Y_train)
    y1_pred=boostmodel.predict( X_train)
    train_errors=1-metrics.accuracy_score(Y_train, y1_pred)
    train_result.append(train_errors)
```

```python
#Plot of train error vs nu,ber of trees
# x axis values
x = parameter_range
# corresponding y axis values
y = train_result
plt.plot(x, y, color='orange', linewidth = 3,marker='o', markerfacecolor='pink', markersize=12)
# plotting the points
plt.plot(x, y)

# naming the x axis
plt.xlabel('Number of trees')
# naming the y axis
plt.ylabel('Train error')

# giving a title to my graph
plt.title('Train Error vs Number of Trees')

# function to show the plot
plt.show()
```
